```python
import pandas as pd
medicalTranscript_df = pd.read_csv("med_transcripts.csv")
medicalTranscript_df.head()
```

## Data Preparation

```python
#check rows with nulls
medicalTranscript_df.isnull().value_counts()
#.sum().sort_values(ascending = False)
```

```python
medicalTranscript_df = medicalTranscript_df[medicalTranscript_df['transcription'].notna()]
medicalTranscript_df.info()
```

```python
#Convert transcript into lowercase
medicalTranscript_df['transcription'] = medicalTranscript_df['transcription'].str.lower()
medicalTranscript_df.head()
```

```python
medicalTranscript_df['transcription'][0]
```

```python
#remove stop words, irrelevant words
#preprocessing part of the code
word_tokenized = nltk.word_tokenize(medicalTranscript_df['transcription'][0])
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
word_tokenized = [word for word in word_tokenized if word not in stopwords]
print (word_tokenized)
```

```python
print(word_tokenized)
```

```python
transcripts = medicalTranscript_df['transcription']
labels = medicalTranscript_df['medical_specialty']
data_with_labels = list(zip(transcripts, labels))
print("data_with_labels[0]: ", data_with_labels[0])

```

## Split the Dataset

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(data_with_labels, test_size = 0.2, random_state = 7)
print("Number of training medical transcripts: ", len(train_set))
print ("train_set[0]: ", train_set[0])
print("Number of testing medical transcripts: ", len(test_set))
```

## Building Unigram feature sets

```python
#Example
wonderful_dict = {"wonderful": True}
awful_dict = {"awful", True}
feature_sets = [(wonderful_dict, "pos"), (awful_dict, "neg")]
print (feature_sets)
```

```python
def unigram_feature(data) -> [[dict,str]]:
    med_transcritps = [d[0] for d in data]
    med_speciality = [d[1] for d in data]
    #feature_dicts = [{word: True for word in nltk.word_tokenize(str(med_transcript))} for med_transcript in med_transcritps]
    feature_dicts = [{word : True for word in nltk.word_tokenize(str(med_transcript))} for med_transcript in med_transcritps]
    return list(zip(feature_dicts, med_speciality))
```

```python
#Example of unigram feature for one of the medical transcritps
debug_feature_sets = unigram_feature(train_set)
debug_sample = debug_feature_sets[10]
print (debug_sample)
#print (debug_feature_sets)
```

## Training Model

```python
debug_feature_sets = unigram_feature(train_set)
classifier = nltk.NaiveBayesClassifier.train(debug_feature_sets)
```

## Prediction and Evaluation

```python
#Prediction
test_feature_sets = unigram_feature(test_set)
test_dict = test_feature_sets[0][0]
print("real value: ", test_feature_sets[0][1])
prediction = classifier.classify(test_dict)
print("predicted value: ", prediction)

#Evalution 
def evaluate_Classifier(classifier, test_data) -> float:
    test_feature_sets = unigram_feature(test_data)
    return nltk.classify.accuracy(classifier, test_feature_sets)

acc = evaluate_Classifier(classifier, test_set)
print("Accuracy: ", acc)
```

```python
num_test_labels = len(test_set)
print("num_test_labels: ", num_test_labels)

true_labels = []
for i in range(num_test_labels):
    true_labels.append(test_feature_sets[i][1])
print("true_labels: ", true_labels)

predicted_labels = []
for i in range(num_test_labels):
    prediction = classifier.classify(test_feature_sets[i][0])
    predicted_labels.append(prediction)
print("predicted_labels: ", predicted_labels)

print(classification_report(true_labels, predicted_labels))
```

## Building Bigram feature sets

```python
def bigram_feature(data) -> [(dict,str)]:
    med_transcripts = [d[0] for d in data]
    med_speciality = [d[1] for d in data]
    feature_dicts = [{word: True for word in nltk.bigrams(nltk.word_tokenize(str(med_transcript)))} for med_transcript in med_transcripts]
    return list(zip(feature_dicts, med_speciality))
```

```python
# Example of bigram feature for one of the medical transcripts
debug_feature_sets = bigram_feature(train_set)
debug_sample = debug_feature_sets[10]
print (debug_sample)
```

## Training the model

```python
bigram_feature_sets = bigram_feature(train_set)
bigram_classifier = nltk.NaiveBayesClassifier.train(bigram_feature_sets)
```

## Prediction and Evaluation

```python
# Prediction
test_feature_sets = bigram_feature(test_set)
test_dict = test_feature_sets[1][0]
print("real value: ", test_feature_sets[1][1])
prediction = bigram_classifier.classify(test_dict)
print("predicted value: ", prediction)

#Evaluate
def evaluate_bigram_classifier(classifier, test_data) -> float:
    test_feature_sets = bigram_feature(test_data)
    return nltk.classify.accuracy(classifier, test_feature_sets)

acc_bigram = evaluate_bigram_classifier(bigram_classifier, test_set)
print(acc_bigram)


```

```python
num_test_labels = len(test_set)
print("num_test_labels: ", num_test_labels)

true_labels = []
for i in range(num_test_labels):
    true_labels.append(test_feature_sets[i][1])
print("true_labels: ", true_labels)

predicted_labels = []
for i in range(num_test_labels):
    prediction = bigram_classifier.classify(test_feature_sets[i][0])
    predicted_labels.append(prediction)
print("predicted_labels: ", predicted_labels)

print(classification_report(true_labels, predicted_labels))
```

## Building Trigram feature set

```python
def trigram_feature(data) -> [(dict,str)]:
    med_transcripts = [d[0] for d in data]
    med_speciality = [d[1] for d in data]
    feature_dicts = [{word: True for word in nltk.trigrams(nltk.word_tokenize(str(med_transcript)))} for med_transcript in med_transcripts]
    return list(zip(feature_dicts, med_speciality))
```
